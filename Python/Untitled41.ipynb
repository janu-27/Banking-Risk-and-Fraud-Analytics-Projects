```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Comprehensive_Banking_Database.csv")
```

```python
df.drop_duplicates(inplace=True)
```

```python
date_cols = [
    'Transaction Date',
    'Feedback Date',
    'Resolution Date',
    'Payment Due Date',
    'Last Transaction Date',
    'Date Of Account Opening'
]

for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')
print(df[col])
```

```python
bins = [18,25,35,45,60,100]
labels = ['18-25','26-35','36-45','46-60','60+']

df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels)
print(df['Age Group'])
```

```python
df['Credit Utilization %'] = (
    df['Credit Card Balance'] /
    df['Credit Limit']
) * 100
print(df['Credit Utilization %'])
```

```python
df['High Risk Customer'] = np.where(
    df['Anomaly'] == 1,
    'Yes',
    'No'
)
print(df['High Risk Customer'])
```

```python
df['Estimated EMI'] = (
    df['Loan Amount'] /
    df['Loan Term']
)
print(df['Estimated EMI'])
```

```python
df.to_csv("cleaned_banking_data.csv", index=False)
```

```python
plt.figure(figsize=(10,5))
plt.hist(df['Transaction Amount'], bins=30)
plt.title('Transaction Amount Distribution')
plt.show()
```

```python
df['Account Type'].value_counts().plot(kind='bar')
plt.title('Account Type Distribution')
plt.show()
```

```python
df['Loan Status'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)
plt.show()
```

```python
branch = df.groupby('Branch ID')['Transaction Amount'].sum()

branch.plot(kind='barh')
plt.show()
```

```python
sns.countplot(x='Anomaly', data=df)
plt.title('Fraud Detection')
plt.show()
```

```python
sns.scatterplot(
    x='Credit Limit',
    y='Credit Card Balance',
    data=df
)
plt.show()
```
